QUESTION 1

1. Power BI Desktop

Power BI Desktop is a Windows-based application used to:

Connect to multiple data sources (Excel, SQL Server, CSV, cloud services, etc.)

Clean and transform data using Power Query

Create data models using DAX

Design interactive reports and visuals

 It is mainly used by report developers and analysts.

2. Power BI Service (PowerBI.com)

Power BI Service is a cloud-based platform where users:

Publish reports created in Power BI Desktop

Create dashboards by pinning visuals

Share reports securely with others

Schedule data refreshes

Collaborate using workspaces and apps

It is used for collaboration, sharing, and governance.

3. Power BI Mobile

Power BI Mobile is a mobile application available on Android, iOS, and Windows that allows users to:

View and interact with reports and dashboards on the go

Receive data-driven alerts

Access real-time insights anytime, anywhere

 Best suited for executives and field users.

4. Power BI Gateway

Power BI Gateway acts as a bridge between on-premises data sources and Power BI Service:

Enables secure data refresh from on-prem databases (SQL Server, Oracle, etc.)

Allows live connections without moving data to the cloud

Supports scheduled and real-time data updates

 Essential for organizations using on-premises data sources.

### **Pie Chart vs Donut Chart**

* **Pie Chart**: Shows parts of a whole in a simple circular form.
* **Donut Chart**: Similar to a pie chart but has a center space for extra information.

**When to prefer**

* Use **Pie Chart** for a simple percentage comparison.
* Use **Donut Chart** when you want to show a total or KPI in the center.

**Example**

* Pie Chart: Market share of different brands.
* Donut Chart: Sales by category with total sales in the center.

---

### **Bar Chart vs Column Chart**

* **Bar Chart**: Uses horizontal bars.
* **Column Chart**: Uses vertical bars.

**When to prefer**

* Use **Bar Chart** for long category names or many categories.
* Use **Column Chart** for time-based or trend analysis.

**Example**

* Bar Chart: Sales by country.
* Column Chart: Monthly sales trend.

1. Star Schema vs Snowflake Schema

Star Schema has one fact table connected directly to dimension tables. It is simple, faster, and best for Power BI reporting.

Snowflake Schema has normalized dimension tables linked to other dimensions. It is more complex and slower in Power BI.

Significance:
Star schema improves performance, simplifies relationships, and makes analysis easier.

2. Primary Key vs Foreign Key (Power BI)

Primary Key uniquely identifies records in a table (usually in dimension tables).

Foreign Key links a table to another tableâ€™s primary key (usually in fact tables).

Significance:
They define relationships and allow correct filtering and aggregation in Power BI.

3. Why Cardinality Is Important

Cardinality defines how tables are related (one-to-many, many-to-many, etc.).

Correct cardinality ensures accurate results, proper filtering, and good performance.

## **Calculated Column vs Measure**

**Calculated Column**

* Calculated at **row level**
* Stored in the table
* Uses **row context**
* Increases model size

**Example:**
`Profit = Sales[Revenue] - Sales[Cost]`

---

**Measure**

* Calculated **at query time**
* Not stored in the table
* Uses **filter context**
* Used in visuals (totals, averages, KPIs)

**Example:**
`Total Sales = SUM(Sales[Revenue])`

---

## **Row Context**

* Calculation happens **row by row**
* Used in calculated columns

**Example:**
Calculating profit for each sales row.

---

## **Filter Context**

* Filters applied by **visuals, slicers, or filters**
* Used in measures

**Example:**
Total Sales for **Year = 2024** selected in a slicer.
